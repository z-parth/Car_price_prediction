```python
import sys
from pathlib import Path
import numpy as np
PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT))
```

```python
import pandas as pd
from src.features.encoded import encode_train_test
```

```python
DATA_DIR = PROJECT_ROOT / "data" / "processed"

train_df = pd.read_csv(DATA_DIR / "train.csv")
test_df = pd.read_csv(DATA_DIR / "test.csv")



X_train, X_test, y_train, y_test = encode_train_test(
    train_df,
    test_df,
    target_col="price"   
)
```

```python
from pathlib import Path
import os

PROJECT_ROOT = Path(r"C:\Users\iamli\OneDrive\Desktop\Pricing\mlruns")  
MLRUNS_DIR = PROJECT_ROOT / "mlruns"

os.environ["MLFLOW_TRACKING_URI"] = f"file:///{MLRUNS_DIR.as_posix()}"
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import mlflow
import mlflow.sklearn
import math

mlflow.set_experiment("car_price_prediction")

with mlflow.start_run(run_name="RandomForest_baseline"):
    model = RandomForestRegressor(
        n_estimators=300,
        max_depth=12,
        random_state=42
    )

    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    mlflow.log_params(model.get_params())
    mlflow.log_metric("MAE", mean_absolute_error(y_test, preds))
    mlflow.log_metric("RMSE", np.sqrt(mean_squared_error(y_test, preds)))
    mlflow.log_metric("R2", r2_score(y_test,preds))
```

```python
from sklearn.linear_model import LinearRegression

with mlflow.start_run(run_name="LinearRegression_baseline"):
    model = LinearRegression()

    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    mlflow.log_params(model.get_params())
    mlflow.log_metric("MAE", mean_absolute_error(y_test, preds))
    mlflow.log_metric("RMSE", np.sqrt(mean_squared_error(y_test, preds)))
    mlflow.log_metric("R2", r2_score(y_test, preds))
```
